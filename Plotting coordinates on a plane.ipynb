```python
x = 3
y = 5

plt.plot(x, y,'ro')

plt.axis('square')
plt.axis([-6, 6, -6,6])
plt.grid()

plt.show()
```

```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2, -3, 4]

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label='point %s' %i)
    

plt.axis('square')
plt.grid()
plt.legend()

plt.show()
```

```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2, -3, 4]

for i in range(0, len(x)):
    plt.plot(x[i], y[i], 'o', label='point %s' %(1 +i))
    

plt.axis('square')
plt.grid()
plt.legend()

plt.show()
```

```python
plt.plot(4, 2, 'r+')

axis = plt.gca()

plt.show()
```

```python
x = sym.symbols('x')
```

```python
y = x**2 - 3 * x

for pnti in range(-11, 12):
    plt.plot(pnti, y.subs(x,pnti), 'o')
    
plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)

plt.show()
```
